Text Summarization is the process of creating a summary of a certain document which contains an important information from that document.

There is a huge amount of data available digitally and it is necessary to develop a unique procedure to immediately summarize a long text while keeping the main idea. Text summarization also makes it possible to shorten the reading time, speed up information searches and obtain as much information as possible on a subject.

The main goal of using machine learning for text summarization is to reduce the reference text to a smaller version while keeping its knowledge alongside its meaning.

We dont need to use a lot of machine learning here. We can easily summarize text without training a model. But still, we need to use some natural language processing for that. We will be using the `NLTK` library. 

```python
import nltk
import string
from heapq import nlargest
```

Now lets perform some steps for removing punctuations from the text. Then we need to do some steps of text processing and at the end, we will simply tokenize the text and see the results for text summarization.

```python
text = """Steve was born in Tokyo, Japan in 1950. He moved to London with his parents 
when he was 5 years old. Steve started school there and his father began work at the hospital.
His mother was a house wife and he had four brothers.

He lived in England for 2 years then moved to Amman, Jordan where he lived there for 10 years.
Steve then moved to Cyprus to study at the Mediterranean University.
Unfortunately, he did not succeed and returned to Jordan. 
His parents were very unhappy so he decided to try in America."""

text.count(".")
```

```python
string.punctuation
```

```python
nopuch =[char for char in text if char not in string.punctuation]
print(nopuch)
```

```python
nopuch = "".join(nopuch)
nopuch
```

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
print(stop_words)
```

```python
processed_text = [word for word in nopuch.split() if word.lower() not in stop_words]
print(processed_text)
```

```python
word_freq = {}
for word in processed_text:
    if word not in word_freq:
        word_freq[word] = 1
    else:
        word_freq[word] = word_freq[word] + 1

print(word_freq)
```

```python
max_freq = max(word_freq.values())

print(max_freq)
```

```python
# proportion of values

for word in word_freq.keys():
    word_freq[word] = (word_freq[word]/max_freq)

print(word_freq)
```

### Tokenize the Text

```python
nltk.download('punkt')
sent_list = nltk.sent_tokenize(text) # It will make a list of text. Every next element will start after full stop
sent_list
```

```python
# nltk.word_tokenize('Then we need to do'.lower()) # It will make a list of word
```

```python
sent_score = {}

for sent in sent_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_freq.keys():
            if sent not in sent_score.keys():
                sent_score[sent] = word_freq[word]
            else:
                sent_score[sent] += word_freq[word]

sent_score
```

```python
summary_sents = nlargest(n =2 ,iterable =  sent_score, key=sent_score.get) # It find the largest elements in a datasets
summary_sents
```

```python
summary = " ".join(summary_sents)
print(summary)
```
